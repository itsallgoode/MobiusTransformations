```python
import pandas as pd
import numpy as np
from numpy import *
import random
from mobius_transformations import MobiusTransform
import re
```

```python
df = pd.read_csv('fulltrain_100.csv')
```

```python
def extract_key(col_name):

    match = re.match(r'(.+)_([XY])$', col_name)
    if match:
        landmark_name = match.group(1)
        coord = match.group(2)
        coord_order = 0 if coord == 'X' else 1  
        return (landmark_name, coord_order)
    else:

        return (float('inf'), col_name)
```

```python
def aggregate_dataframe(df):
    return df.agg(lambda x: '[' + ','.join(x.astype(str)) + ']').to_frame().transpose()
```

```python
def unpack_df(df):

    to_skip = ['video_fps', 'video_size_height', 'video_size_width', 'labels']

    dataframes = []

    for index, row in df.iterrows():
        curr_row = {}
        
        for column in row.index:
            if column in to_skip:
                continue
            values = row[column].strip('[]').split(',')
            curr_row[column] = np.array(values, dtype=float)

        new_df = pd.DataFrame(curr_row)
        columns = new_df.columns.tolist()
        sorted_columns = sorted(columns, key=extract_key)
        new_df = new_df.reindex(columns=sorted_columns)

        if random.random() <= 0.25:

            transformer = MobiusTransform(new_df)
            transformed_df = transformer.transform()
            transformed_df = aggregate_dataframe(transformed_df)
            transformed_df['video_fps'] = row['video_fps']
            transformed_df['video_size_height'] = row['video_size_height']
            transformed_df['video_size_width'] = row['video_size_width']
            transformed_df['labels'] = row['labels']
            dataframes.append(transformed_df)

        new_df = aggregate_dataframe(new_df)
        new_df['video_fps'] = row['video_fps']
        new_df['video_size_height'] = row['video_size_height']
        new_df['video_size_width'] = row['video_size_width']
        new_df['labels'] = row['labels']
        dataframes.append(new_df)
    
    return dataframes
```

```python
unpacked_data = unpack_df(df)
new_df = pd.concat(unpacked_data)
new_df.to_csv('M11P25limited.csv', index=False)
```
